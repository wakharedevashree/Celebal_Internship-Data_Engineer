## Step 3: Handle Missing Values

### 3.1 Identify Missing Values


```python
# Count of missing values per column
missing_counts = df.isnull().sum()

```

```python
# Percentage of missing values per column
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_summary = pd.DataFrame({
    'missing_count': missing_counts,
    'missing_pct': missing_pct
})


missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_count', ascending=False)
```

```python
# Make a working copy so the original df is preserved for reference
df_clean = df.copy()

# Fill numeric column
df_clean['discount'] = df_clean['discount'].fillna(0)


fill_values = {
    'seller_name': 'Unknown Seller',
    'seller_information': 'Not Available',
    'variations': 'No Variation',
    'what_customers_said': 'No Review Text',
    'videos': 'No Video'
}
df_clean = df_clean.fillna(value=fill_values)


df_clean[['discount', 'seller_name', 'seller_information', 'variations',
          'what_customers_said', 'videos']].isnull().sum()
```

```python

print("Remaining missing values across ALL columns:")
print(df_clean.isnull().sum().sum())
```

## Step 4: Perform Basic Operations (Filter Rows, Select Columns)


```python
# Select a subset of useful columns for analysis
cols_of_interest = ['product_id', 'title', 'category', 'rating',
                     'ratings_count', 'initial_price', 'discount', 'final_price']

df_subset = df_clean[cols_of_interest]
df_subset.head()
```

```python
# Filter: highly rated products (rating >= 4.5)
high_rated = df_clean[df_clean['rating'] >= 4.5]
print(f"Products with rating >= 4.5: {len(high_rated)}")
high_rated[['title', 'category', 'rating']].head()
```

```python
# Filter: products in a specific category
tshirts = df_clean[df_clean['category'] == 'tshirts']
print(f"Number of t-shirt products: {len(tshirts)}")
tshirts[['title', 'category', 'rating', 'final_price']].head()
```

```python
# Filter: popular products (ratings_count > 100) combined with high rating
popular_and_good = df_clean[(df_clean['ratings_count'] > 100) & (df_clean['rating'] >= 4.0)]
print(f"Popular AND well-rated products: {len(popular_and_good)}")
popular_and_good[['title', 'category', 'rating', 'ratings_count']].sort_values(
    'ratings_count', ascending=False).head(10)
```

## Step 5: Remove Duplicates


```python
# Check for fully duplicated rows
full_dupes = df_clean.duplicated().sum()
print(f"Fully duplicated rows: {full_dupes}")

# Check for duplicate product_id (the natural unique identifier)
id_dupes = df_clean.duplicated(subset=['product_id']).sum()
print(f"Duplicate product_id rows: {id_dupes}")
```

```python
# Remove duplicates (keeps the first occurrence)
before = len(df_clean)
df_clean = df_clean.drop_duplicates()
df_clean = df_clean.drop_duplicates(subset=['product_id'], keep='first')
after = len(df_clean)

print(f"Rows before removing duplicates: {before}")
print(f"Rows after removing duplicates:  {after}")
print(f"Duplicate rows removed: {before - after}")
```

## Step 6: Create a Derived Column — `total_amount = price * quantity`


```python
# --- 6.1 Clean the price column into numeric ---
df_clean['final_price_clean'] = (
    df_clean['final_price']
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
)
df_clean['final_price_clean'] = pd.to_numeric(df_clean['final_price_clean'], errors='coerce')

# initial_price is already numeric
df_clean['initial_price_clean'] = pd.to_numeric(df_clean['initial_price'], errors='coerce')


df_clean['price'] = df_clean['final_price_clean']

print("Any NaNs created while cleaning price?", df_clean['price'].isnull().sum())
df_clean[['final_price', 'final_price_clean', 'initial_price', 'price']].head()
```

```python
#6.2 Simulate a 'quantity' column (units ordered per product) ---
# NOTE: This dataset has no real order-level 'quantity' field.
# We simulate realistic order quantities (1 to 5 units) using a fixed random seed
np.random.seed(42)
df_clean['quantity'] = np.random.randint(1, 6, size=len(df_clean))  # random int between 1 and 5

df_clean[['title', 'price', 'quantity']].head()
```

```python
# --- 6.3 Create the derived column: total_amount = price * quantity ---
df_clean['total_amount'] = df_clean['price'] * df_clean['quantity']

df_clean[['title', 'category', 'price', 'quantity', 'total_amount']].head(10)
```

```python

print("total_amount summary statistics:")
df_clean['total_amount'].describe()
```

```python
# 3.3 Avg Total Amount by Category and Rating Band
df_clean['rating_band'] = pd.cut(df_clean['rating'], bins=[-0.1, 2.5, 4.0, 5.0],
                                  labels=['Low (0-2.5)', 'Medium (2.5-4)', 'High (4-5)'])
top6 = df_clean['category'].value_counts().head(6).index
df_top6 = df_clean[df_clean['category'].isin(top6)]
pivot = df_top6.pivot_table(values='total_amount', index='category', columns='rating_band',
                             aggfunc='mean', observed=True)
fig, ax = plt.subplots(figsize=(10, 6))
pivot.plot(kind='bar', ax=ax, colormap='plasma')
ax.set_title('Average Total Amount by Category and Rating Band', fontsize=14, fontweight='bold')
ax.set_xlabel('Category')
ax.set_ylabel('Average Total Amount (₹)')
ax.legend(title='Rating Band')
plt.xticks(rotation=40, ha='right')
plt.savefig( '11_multivariate_category_ratingband_totalamount.png', bbox_inches='tight', dpi=150)
plt.show()
```

## Summary

**Brief Summary of Work Done:**

1. **Loaded** `Combined_dataset.csv` into a Pandas DataFrame — 1000 rows × 24 columns of e-commerce product data.
2. **Explored** the dataset using `.head()`, `.tail()`, `.shape`, `.columns`, `.dtypes`, `.info()`, and `.describe()` to understand its structure. Found a mix of numeric columns (price, rating, ratings_count) and many text/JSON-style columns (images, specifications, offers).
3. **Identified missing values** in 6 columns (`discount`, `seller_name`, `seller_information`, `variations`, `what_customers_said`, `videos`) ranging from ~12% to ~78% missing. Filled numeric `discount` with `0` and filled text columns with descriptive placeholders rather than dropping rows, since the core analytical fields had no missing data.
4. **Performed basic operations**: selected a focused subset of columns, and filtered rows by conditions (high-rated products, a specific category, and popular + well-rated products using combined conditions).
5. **Removed duplicates**: checked for both fully duplicated rows and duplicate `product_id` values (none were found in this dataset — confirming each product appears exactly once).
6. **Created a derived column** `total_amount = price * quantity`:
   - Cleaned `final_price` from a text/currency format (e.g. `"₹3,995.00"`) into a proper numeric column.
   - Since this is product catalog data (not order data), simulated a reproducible `quantity` column (1–5 units) to demonstrate the calculation.
   - Computed `total_amount` for every product.
7. **Saved** the cleaned dataset as two CSV files:
   - `cleaned_dataset_full.csv` — all original columns plus cleaned/derived columns.
   - `cleaned_dataset.csv` — a focused, analysis-ready table with key business columns only.

**Key takeaway:** Real-world datasets rarely match an assignment's exact expected schema (e.g. no `quantity` column here). The right approach is to understand what's actually available, clean it properly, and clearly document any assumptions or simulated values used to meet requirements — exactly what was done above.
